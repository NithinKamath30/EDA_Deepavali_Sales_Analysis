# Bar graph:

```python
# Group data
occupation_gender = df.groupby(['Occupation', 'Gender']).size().unstack()

# Define x-axis positions and bar heights
xaxis = np.arange(len(occupation_gender))
yaxis_male = occupation_gender['M']
yaxis_female = occupation_gender['F']

# Plot bars
plt.bar(xaxis - 0.2, yaxis_male, width=0.4, label='Male')
plt.bar(xaxis + 0.2, yaxis_female, width=0.4, label='Female')

# Add labels and title
plt.title('Gender Distribution by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.xticks(xaxis, occupation_gender.index, rotation=90)
plt.legend()
plt.show()
```

## Histogram

```python
#bin-Specifies how many bins the data should be divided into.
# rwidth-Controls the relative width of the bars in the histogram.
plt.hist(df['Age'],bins=10,rwidth=0.95)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age distribution')
```

# pie chart

```python
#autopct-Displays percentages on the pie chart
# startangle- Rotates the chart to make it easier to read.

zone_counts = df['Zone'].value_counts()
plt.pie(zone_counts, labels=zone_counts.index, autopct='%1.1f%%', startangle=180, colors=['salmon', 'lightgreen', 'lightblue'])
plt.title('Zone Distribution')
plt.show()
```

# counterplot

```python
# Visualize the count of males and females.
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Gender', hue='Gender', palette='Set2', legend=True)
plt.title('Gender Distribution')
plt.show()
```

## Boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a boxplot for 'Age Group' vs 'Orders', assigning 'Age Group' to 'hue'
sns.boxplot(data=df, x='Age Group', y='Orders', hue='Age Group', palette='muted')

# Add title and labels
plt.title('Orders Distribution by Age Group')
plt.show()
```

## Scatterplot using seaborn    

```python
sns.scatterplot(data=df, x='Orders', y='Amount', hue='Gender', palette='Set2')
plt.title('Scatter Plot: Orders vs Amount by Gender')
plt.show()
```

## Lineplot

```python
#Compare spending habits across different occupations.
sns.lineplot(data=df, x='Occupation', y='Amount', marker='v', color='orange')
plt.title('Occupation vs Amount')
plt.xlabel('Occupation')
plt.ylabel('Amount')
plt.xticks(rotation=90)
plt.legend()
plt.show()

```
